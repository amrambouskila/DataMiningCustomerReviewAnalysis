```python
import numpy as np
import pandas as pd
from textblob import TextBlob
```

```python
df = pd.read_csv(r'./data/1429_1.csv')
df
```

```python
gifts = 0
jobs = 0
returns = 0
for i in range(len(df)):
    text = str(df.loc[i, 'reviews.text']).lower()
    if 'gift' in text:
        gifts += 1
    if 'work' in text or 'job' in text:
        jobs += 1
    if 'return' in text or 'exchange'  in text:
        returns += 1
```
